### Time Complexity

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Time Complexity Graph 그리기
x = np.linspace(0.001, 30, num=30, dtype=float)
```

```python
# Time Coplexity functions
def O_1(n):
    return np.linspace(10, 10, num = len(x))

def O_logn(n):
    return np.log(n)

def O_N(n):
    return n

def O_nlogn(n):
    return n * np.log(n)

def O_n2(n):
    return n**2

def O_2_to_the_n(n):
    return 2**n

def O_factorial(n):
    tmp = []

    def factorial(x):
        x = int(x)
        lst = range(1, x+1)
        result = 1
        for k in lst:
            result *= int(k)
        return result
    
    for x in n:
        result = factorial(x)
        tmp.append(result)

    return np.array(tmp)
```

```python
# plot
plt.figure(figsize=(12,8))
plt.title("Time Complexity O(N)")

plt.plot(x, O_1(x), label = "O(k)")
plt.plot(x, O_logn(x), label = "O(logn)")
plt.plot(x, O_nlogn(x), label = "O(nlogn)")
plt.plot(x, O_n2(x), label = "O(n$^2$)")
plt.plot(x, O_2_to_the_n(x), label = "O(2$^n$)")
plt.plot(x, O_factorial(x), label = "O(n!)")

plt.yscale("log")
plt.legend(fontsize= 20)
plt.xlim(0, 20)
plt.ylim(0.001, 1.e7)
```
